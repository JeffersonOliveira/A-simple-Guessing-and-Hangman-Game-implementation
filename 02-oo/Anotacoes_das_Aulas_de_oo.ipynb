# Anotações das Aulas de Orientação a Objetos em Python 3

## Entendendo o o conceito de objetos na prática com exemplo de conta bancária:

```python
# Declarando os valores soltos de uma conta
numero = 123
titular = "Nico"
saldo = 55.0
limite = 1000.0
```

```python
# Mas a melhor forma de fazer isso é através do uso de dicionários em python:
conta = {'numero' : 123, 
         'titular' : 'Nico',
         'saldo' : 55.0,
         'limite' : 1000.0}

conta
```

```python
conta['numero']
```

```python
conta2 = {'numero' : 321, 
         'titular' : 'Marco',
         'saldo' : 110.0,
         'limite' : 1000.0}

conta2
```

```python
for c in [conta,conta2]:
  print(c)
```
